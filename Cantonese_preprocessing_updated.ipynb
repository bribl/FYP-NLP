```python
stop_words = pycantonese.stop_words()

stop_words= list(stop_words)
my_stopwords = ['我','哋','應該']  #手動加入stop words
stop_words += my_stopwords
stop_words
```

```python
import pandas as pd
df = pd.read_csv('merged_updated.csv')


df = df[df['Category'] != 'None']

df
```

```python
# 提取廣東話內容的columns

df_content = df[['Title','Main_content','Comments']]
df_content.head(10)
```

```python
import re

def preprocess_text(text):
    if isinstance(text, str):  # 檢查 text 是否為字符串類型
        text = re.sub(r'[^\w\s]', '', text)
        text = text.replace('\n', '')
        
        # Stop Word Removal
        stop_words = []  # Stop word list, customized according to needs
        words = [word for word in text if word not in stop_words]
        
        # Word Segmentation
        words = list(text)
        
        # Text Cleaning
        words = [re.sub(r'[^\u4e00-\u9fff\s]', '', word) for word in words]
        
        # Lowercasing
        words = [word.lower() for word in words]
        
        # Return preprocessed text
        processed_text = ' '.join(words)
        return processed_text
    else:
        return ''  # 非字符串類型返回空字符串
    
# apply function
df_content['Title'] = df_content['Title'].apply(preprocess_text)
df_content['Main_content'] = df_content['Main_content'].apply(preprocess_text)
df_content['Comments'] = df_content['Comments'].apply(preprocess_text)
```

```python
df_content
```

```python
## 去除廣東話常用詞 (stop words)

df_content['Title'] = df_content['Title'].astype(str)
df_content['Main_content'] = df_content['Main_content'].astype(str)
df_content['Comments'] = df_content['Comments'].astype(str)

df_content['Title'] = df_content['Title'].apply(lambda x: ' '.join([word for word in re.split('\W+', x) if word.lower() not in stop_words]))
df_content['Main_content'] = df_content['Main_content'].apply(lambda x: ' '.join([word for word in re.split('\W+', x) if word.lower() not in stop_words]))
df_content['Comments'] = df_content['Comments'].apply(lambda x: ' '.join([word for word in re.split('\W+', x) if word.lower() not in stop_words]))
```

```python
df_content
```

```python
# 同時為句子進行分詞和加入拼音

import pandas as pd
from pycantonese import characters_to_jyutping


# 在指定列上應用characters_to_jyutping函數
df_content['Main_content'] = df_content['Main_content'].apply(lambda x: characters_to_jyutping(x) if isinstance(x, str) else x)
df_content['Title'] = df_content['Title'].apply(lambda x: characters_to_jyutping(x) if isinstance(x, str) else x)
df_content['Comments'] = df_content['Comments'].apply(lambda x: characters_to_jyutping(x) if isinstance(x, str) else x)
```
