```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.firefox.options import Options
import time
import pandas as pd

def scrape_lihkg_data(start_page, end_page):
    options = Options()
    options.headless = True

    driver = webdriver.Firefox(options=options)

    data = []

    for post_id in range(start_page, end_page + 1):
        url = f"https://lihkg.com/thread/{post_id}/page/1" 
        driver.get(url)
        time.sleep(1)

        category = driver.find_elements(By.CLASS_NAME, "ozrBNVWlaHP7dx-mUf7p2")
        category_text = [cate.text for cate in category]

        title = driver.find_elements(By.CLASS_NAME, "CrheYfsiQFY-vLMnO378W")
        title_text = [t.text for t in title]

        content = driver.find_elements(By.CLASS_NAME, "_2cNsJna0_hV8tdMj3X6_gJ")
        #content_combined = " ".join([c.text for c in content])
        content_combined = [c.text for c in content]

        element = driver.find_element(By.TAG_NAME, 'a')
        href = element.get_attribute('href')
        number = href.split('/')[-1]

        data.append({'Category': category_text, 'Number': number, 'Title': title_text, 'Comment': content_combined, 'URL': url, 'post_id': post_id})

    driver.quit()

    df = pd.DataFrame(data)
    return df
```

```python
def merge_range_of_dataframes(start_index, end_index):
    dfs = []
    for i in range(start_index, end_index + 1):
        dfs.append(globals()[f"df{i}"])  

    merged_df = pd.concat(dfs, ignore_index=True)  
    return merged_df
merged_df = merge_range_of_dataframes(1, 8)
merged_df
```
