```python
## 關鍵字提取

from googleapiclient.discovery import build
from googleapiclient.errors import HttpError
from sklearn.feature_extraction.text import CountVectorizer

# Get video information (parameters can be customized as needed)
try:
    response = youtube.search().list(
        part="snippet",
        q="周杰倫",  # 輸入為您想要搜尋的查詢詞
        maxResults=10  # 取得的影片數量
    ).execute()

    # Extract keywords
    video_texts = []
    for item in response["items"]:
        video_title = item["snippet"]["title"]
        video_description = item["snippet"]["description"]
        video_tags = item["snippet"]["tags"] if "tags" in item["snippet"] else []

        
        video_text = video_title + " " + video_description + " " + " ".join(video_tags)
        video_texts.append(video_text)

    # Feature extraction
    vectorizer = CountVectorizer()
    features = vectorizer.fit_transform(video_texts)

    # Get the extracted keywords
    keywords = vectorizer.get_feature_names()
    print("Extracted keywords：", keywords)

except HttpError as e:
    print("An error occurred：", e)
```

```python
## 評論情感分析 (這個預測不太準確而且只能預測英文的評論) (Sentiment:[負面情感, 中性情感, 正面情感], Subjectivity:[客觀,主觀] )
from googleapiclient.discovery import build
from textblob import TextBlob

video_url = "https://www.youtube.com/watch?v=09R8_2nJtjg"  # 輸入你想分析的影片連結
parsed_url = urlparse(video_url)
video_id = parse_qs(parsed_url.query)['v'][0]

response = youtube.commentThreads().list(
    part='snippet',
    videoId=video_id,
    textFormat='plainText',
    maxResults=100
).execute()

# Extract review text and conduct sentiment analysis
for item in response['items']:
    comment = item['snippet']['topLevelComment']['snippet']['textOriginal']
    blob = TextBlob(comment)
    sentiment = blob.sentiment
    print('Comment:', comment)
    print('Sentiment:', sentiment.polarity)  # Sentiment, range is [-1, 1]
    print('Subjectivity:', sentiment.subjectivity)  # Subjectivity, range is [0, 1]
    print('---')
```

```python
# 使用jieba進行中文分詞 (jieba是常用的中文分詞庫)
import jieba
from urllib.parse import urlparse, parse_qs
import googleapiclient.discovery



video_url = "https://www.youtube.com/watch?v=gAotCGoKhD0"  # 輸入你想分析的影片鏈接
parsed_url = urlparse(video_url)
video_id = parse_qs(parsed_url.query)['v'][0]

response = youtube.commentThreads().list(
    part='snippet',
    videoId=video_id,
    textFormat='plainText',
    maxResults=100 # 提取評論的數量
).execute()

# 使用jieba進行分詞
for item in response['items']:
    comment = item['snippet']['topLevelComment']['snippet']['textOriginal']
    seg_list = jieba.cut(comment)
    seg_result = " ".join(seg_list)
    print('原始評論：', comment)
    print('分詞結果：', seg_result)
    print('---')
```
