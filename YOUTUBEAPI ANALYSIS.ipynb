```python
## è©•è«–æƒ…æ„Ÿåˆ†æ (é€™å€‹é æ¸¬ä¸å¤ªæº–ç¢ºè€Œä¸”åªèƒ½é æ¸¬è‹±æ–‡çš„è©•è«–) (Sentiment:[è² é¢æƒ…æ„Ÿ, ä¸­æ€§æƒ…æ„Ÿ, æ­£é¢æƒ…æ„Ÿ], Subjectivity:[å®¢è§€,ä¸»è§€] )
from googleapiclient.discovery import build
from textblob import TextBlob

video_url = "https://www.youtube.com/watch?v=09R8_2nJtjg"  # è¼¸å…¥ä½ æƒ³åˆ†æçš„å½±ç‰‡é€£çµ
parsed_url = urlparse(video_url)
video_id = parse_qs(parsed_url.query)['v'][0]

response = youtube.commentThreads().list(
    part='snippet',
    videoId=video_id,
    textFormat='plainText',
    maxResults=100
).execute()

# Extract review text and conduct sentiment analysis
for item in response['items']:
    comment = item['snippet']['topLevelComment']['snippet']['textOriginal']
    blob = TextBlob(comment)
    sentiment = blob.sentiment
    print('Comment:', comment)
    print('Sentiment:', sentiment.polarity)  # Sentiment, range is [-1, 1]
    print('Subjectivity:', sentiment.subjectivity)  # Subjectivity, range is [0, 1]
    print('---')
```
