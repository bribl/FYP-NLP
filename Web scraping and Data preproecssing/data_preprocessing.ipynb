```python
import pycantonese
import pandas as pd
import numpy as np 
from pycantonese.word_segmentation import Segmenter
import re

stop_words = pycantonese.stop_words()
segmenter = Segmenter(max_word_length=3)
#Updated text preprocessing function
def preprocess_text(text):
    # Check if it is a string
    if not isinstance(text, str):
        # If it is not a string, you can choose to skip processing and return an empty string.
        
        return ''
    # Remove English letters and special symbols, leaving only Chinese characters
    text = re.sub(r'[^\u4e00-\u9fff]', '', text)
    #Word segmentation processing
   
    words = pycantonese.segment(text, cls=segmenter)
    # Remove stop words
    words = [word for word in words if word not in stop_words]
    
    # Reassemble words into strings
    processed_text = ' '.join(words)
    
    return processed_text


df['Comments_preprocessed'] = df['Comments'].apply(preprocess_text)


#print(df['Comment_preprocessed','Comments'])
df[['Comments_preprocessed','URL','Category','Classification']]
```

```python
def check_stop_words(column):
    stop_words = pycantonese.stop_words()  
    for word in column:
        if word in stop_words:
            print(word)
        else:
            print("Stop words not found")

column_to_check = df['Comments_preprocessed']  
check_stop_words(column_to_check)
```

```python
df[['Comments_preprocessed','URL','Category','Classification']].to_csv("preprocessed_data.csv",encoding='utf-8-sig',index=False)
```
