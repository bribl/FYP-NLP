```python
import pandas as pd
df = pd.read_csv('廣東話.csv')
df1 = pd.read_csv('posts.csv')
```

```python
df_comment = df[['Comment_content']]

df1_comment = df1[['Title','Main_content','Category']]
```

```python
import re

def preprocess_text(text):
    text = re.sub(r'[^\w\s]', '', text)
    text = text.replace('\n', '')
    # Word Segmentation
    words = list(text)
    
    # Stop Word Removal
    stop_words = []  # Stop word list, customized according to needs
    words = [word for word in words if word not in stop_words]
    
    # Text Cleaning
    words = [re.sub(r'[^\u4e00-\u9fff\s]', '', word) for word in words]
    
    # Lowercasing
    words = [word.lower() for word in words]
    

    
    # Return preprocessed text
    processed_text = ' '.join(words)
    return processed_text

# apply function
df_comment['Comment_content'] = df_comment['Comment_content'].apply(preprocess_text)
```

```python
# Fill null values into empty string
df1_comment['Title'] = df1_comment['Title'].fillna('')
df1_comment['Main_content'] = df1_comment['Main_content'].fillna('')

# Preprocess Title and Main_content in df1_comment
df1_comment[['Title', 'Main_content']] = df1_comment[['Title', 'Main_content']].applymap(preprocess_text)
```
